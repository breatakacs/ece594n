## I. Introduction

Neurodegenerative diseases impact millions of people worldwide and occur when cerebral nerve cells lose function over time and ultimately die (Burton et al 2004). They often strike later in life and have overlapping symptoms such as cognitive impairment. In conjunction with patients' more complex medical histories, it can be difficult to distinguish and diagnose these disorders.

Parkinson's and Alzheimer's disease are the most common neurodegernative diseases. Symptomatically speaking, Alzheimer's is associated with continuous degeneration of cognitive function, comprehension, judgement, memory and self-control (Yang et al 2019). Parkinson's disease starts with motor dysfunction that can progress to one's cognition (Zarei et al 2012).

Like all neuodegenerative diseases, PD and AD are associated with atrophy in grey matter which is composed of neural cell bodies, axon terminals, and dendrites. Studies implicate α-synuclein aggregates in these diseases at varying degrees, but being able to understand how these conditions specifically impact the structure and shape of the brain may lend incredible insight to doctors when diagnosing patients. 

### Relevant Vocabulary
Brain atrophy: 
- reduction in cerebral tissue

Cortical thickness: 
- Measures the width of gray matter of the human cortex and is often calculated using T1-weighted MRI images
- distance from the interface between cerebrospinal fluid and grey matter to the interface between the white matter cortex and the grey matter cortex

Gray matter density: 
- Measure of gray matter that naturally increases from youth to adulthood
- Calculated as cubic millimeters of gray matter per voxel; unitless

Dementia Lewy body:
- Second most common type of progressive dementia after Alzheimer's disease
- Lewy bodies are aggregates of misfolded alpha synuclein 

## II. Literature Review

### Yang et al, Study of brain morphology change in Alzheimer’s disease and amnestic mild cognitive impairment compared with normal controls, General Psychiatry, 2019


Differences in cortical thickness signified by blue areas in both left and right hemispheres
B and E: Mild cognitive impairment vs normal controls
D and F: Alzheimer's disease vs mild cognitive impairment

Summary:
- Yang et al 2019 investigated how to distinguish early versus late stages of Alzheimer's disease; more specifically how effective two cerebral features, cortical thickness and surface area, are in detecting early AD. The authors hypothesized that patients with Alzheimer's disease and mild cognitive impairment (aMCI) would exhibit significant alterations in cortical thickness and surface area compared to the control group. To study this, MRI images were taken from 20 patients diagnosed with AD, 21 with mild cognitive impairment (aMCI), and 25 set as age-matched and sex-matched normal controls (NC). Surface-based measurements of brain cortical thickness and surface area were obtained from 36 different brain regions in the MRI images. AD patients showed a decrease in the cortical thickness of both hemispheres in 17/36 regions as well as in surface area in 3/36 regions whereas aMCI showed slight variations. This indicates that cortical thickness and surface area are better indicators of progressed AD at later stages of the disease.

Relation to topic
- This study showed that the condition, Alzheimer's disease, is correlated with 2 structural features of brain shape: brain cortical thickness and surface area. It answers the question by showing that AD is more correlated with a reduction in cortical thickness and surface area in particular cerebral regions than other less advanced brain pathologies such as aMCI. This includes wide regions of the frontal, temporal and occipital lobes of the brain.

Relation to Zarei et al 2012
- Both Yang et al 2019 and Zarei et al 2012 look at how cortical thickness relates to specific brain pathologies. Yang and Zarei investigated Alzheimer's disease and Parkinson's disease, respectively. They analyze changes localized in the frontal, temporal, and occipital lobes. These papers identify some affected subregions in both AD and PD, such as lateral occipital cortex and posterior parietal cortices. Fig 1 and 3 in this literature review provide a visual comparison of regions and their associated levels of cortical thinning. 

Relation to Frisoni et al 2002:
- Both Yang et al 2019 and Zarei et al 2002 investigate Alzheimer's disease but use two different morphometric features. Yang uses cortical thickness while Zarei uses gray matter density. In addition, Yang compare various stages of AD while Frisoni just analyzes mild AD. They found some overlap in atrophy in regions including the parahippocampal gyrus and middle temporal gyrus which can be visualized in Fig 1 and 2 of this lit review.

### Frisoni et al, Detection of grey matter loss in mild Alzheimer’s disease with voxel based morphometry, Neurol Neurosurg, 2002


Correlation of disease severity calculated via mini-mental state examination with grey matter density in AD patients

Summary:
- Frisoni et al 2002 aimed to understand how gray matter density relates to neurodegenetation in Alheimer's disease. Based on other studies, it was hypothesized that AD patients would exhibit atrophy of head body and tail of the hippocampus, amygdala, insula, lateral temporal and cingulate cortex, precuneus, and caudate nucleus. To investigate this, MRI images were obtained from 29 patients with AD and 26 healthy controls. Voxel based image processing was done to measure gray matter structures in 27 sub-regions of the forebrain, amygdala, hippocampus, and neocortex. In comparison to the controls, all AD patients showed large regions of atrophy in the hippocampus and amygdalar complex in both hemispheres. All localised regions hypothesized to exhibit atrophy did, and 2 regions novelly involved are uncus and fusiform gyrus. Through this, the authors show grey matter density can be used to assess disease severity and cognitive impairment in patients with AD.

Relation to topic
- This study showed that the condition, Parkinson's disease, is correlated with a structural feature of brain shape: grey matter density. It answers the topic question by showing AD is strongly correlated with a reduction in grey matter density in cerebral surface regions, particularly the hippocampus and amygdala.

Relation to Frisoni et al 2002
- Frisoni et al 2002 and Yang et al 2019 both investigated Alzheimer's disease. They differ in the brain shape feature used to assess atrophy with Frisoni employing gray matter density and Yang with cortical thickness. These two papers overlap in some of the identified atrophied areas such as in the hippocampus and middle temporal gyrus. 

### Zarei et al, Cortical thinning is associated with disease stages and dementia in Parkinson’s disease, Movement Disorders, 2012


Differences in cortical thickness signified by blue areas in both left and right hemispheres 
A: Control vs moderate Parkinson's
B: Control vs Parkinson's with dementia
C: Moderate Parkinson's vs Parkinson's with dementia

Summary:
- Zarei et al 2012 investigated the relationship between cortical thinning and different stages of Parkinson's disease (PD). It was thought that cortical thickness would be altered in a manner dependent upon the severity of PD in consistency with Braak's hypothesis of PD pathological progression. To investigate this, MRI brain images were aquired from ~60 patients diagnosed with either early, moderate, or late-stage PD, as well as from ~40 young and old healthy controls (YHC and OHC). Calculated cortical thickness was correlated with motor and cognitive skills. Compared to contols, average cortical thickness in early PD remained unchanged but was  prominently reduced in the right hemisphere of moderate PD and more extensively in both hemispheres of advanced PD. While no major correlations were observed between cortical thickness and motor skills, regions impacted by cortical thinning changed with cognitive function. In moderate PD, thinning was initially observed in bilateral supramarginal bilateral premotor, bilateral dorsolateral prefrontal, and right inferior-medial temporal cortex. In advanced dementia PD, thinning exapnded to adjacent regions including dorsal and medial frontal cortex as well as the temporal cortex and posterior parietal cortex. This data was consistent with Braak's hypothesis on the pattern of ascending cortical involvement in PD progression and shows that cortical thickness can be used to assess disease severity and cognitive impairment in patients with PD.

Relation to topic
- This study showed that the condition, Parkinson's disease, is correlated with a structural feature of brain shape: brain cortical thickness. It answers the topic question by showing a strong correlation between cortical thickness and advancement of PD.  Regions include medial frontal (premotor and supplementary motor cortex), posterior cingulate, precuneus, lateral occipital, temporal and dorsolateral prefrontal cortex.

Relation to Blair et al 2019
- Zarei et al 2012 and Blair et al 2019 both explored atrophy in Parkinson's disease but with different brain shape features. Zarei used cortical thickness, a vector quantity, and Blair used gray matter density, a scalar measurement. They both heavily analyzed subcortical structures. Blair showed that cortical thickness changes in the hippocampi and amygdala were only significantly noticeable in late-stages of PD. In comparison, Zarei showed gray matter density changes in these regions was noticeable in earlier stages of PD.

### Blair et al, Brain MRI Reveals Ascending Atrophy in Parkinson's Disease Across Severity, Frontiers in Neurology, 2019


Gray matter density differences and relationship with disease duration in amygdala brain regions

Summary:
- Blair et al 2019 aimed to understand how gray matter density relates to neurodegenetation in various stages of Parkinson's disease. Based on the ascending spread of lewy bodies in PD, the authors hypothesized that patients with more advanced PD would show decreases in GMD in subcortical regions compared to cortical regions. To investigate this, MRI images were obtained from ~225 patients with early PD, ~130 with advanced PD, and ~100 sex- and age-matched controls. Voxel based image processing was done to measure gray matter structures in 27 sub-regions of the forebrain, amygdala, hippocampus, and neocortex. Calculated GMD was correlated with cognitive function. Compared to healthy controls, all stages of PD showed a decrease in gray matter density in all subcortical regions analyzed. Furthermore, advanced PD patients showed a widespread reduction of GMD in basal forebrain and amygdala compared to both early PD and healthy controls. Surprisingly, GMD in hippocampal and neo-cortical regions were not impacted by PD. This data indicates an "ascending" pattern of gray matter atrophy when comparing groups differentiated by PD duration.

Relation to topic
- This study showed that the condition, Parkinson's disease, is correlated with a structural feature of brain shape: gray matter density. It answers the question by showing that PD is more correlated with a reduction in gray matter density in subcortical regions of the brain than other less advanced stages of PD. 

How it relates to other papers
- Blair et al 2019 compares to Frisoni et al 2002  in that both use similar methodologies to analyze the same brain feature in different neurodegenerative diseases: PD and AD, respectively. They employ voxel based morphometry to calculate gray matter density. They identified some overlap in regions affected by atrophied gray matter between AD and PD such as the hippocampus/amygdala. 

### Burton et al, Cerebral atrophy in Parkinson's disease with and without dementia: a comparison with Alzheimer's disease, dementia with Lewy bodies and controls, Brain, 2004


Comparison of gray matter volume in Alzheimer's, Parkinson's, and dementia with Lewy bodies

Summary:
- Burton et al 2004 aimed to characterize structural similarities and differences in cerebral atrophy of patients with various neurodegenerative disorders: PD, PD with dementia (PDD), AD, and dementia with Lewy bodies (DLB). The authors hypothesized that brain atrophy would be at similar levels in PDD and DLB than PD or DLB. To investigate this, MRI scans were obtained from 26 PDD, 31 PD, 28 AD, and 17 DLB patients. Voxel-based morphometry was used to investigate loss in grey matter volume (GMV). The study found that PD without dementia involves grey matter loss in frontal areas which extends to the temporal, parietal and subcortical areas in PDD patients. In contrast, AD patients exhibited atrophy of the temporal lobe, including the hippocampus and parahippocampal gyrus. In support of their hypothesis, PDD and DLB had most similar atrophy patterns compared to PD and AD. 

Relation to topic
- This study compared cerebral atrophy in different neurodegenerative diseases using the structural feature of brain shape: gray matter volume. More specifically, it investigates Alzheimer's disease and Parkinson's disease which are the focus of this literature review and then compares them to another disease, dementia with Lewy body. 

- Rather than show that one of these pathologies is more correlated with brain shape overall, this study implicates that PD, PDD, and AD are all associated with reductions in gray matter volume just at different brain regions and levels. The data also refutes claims that PDD and AD show similar cerebral changes, and instead provide evidence that PDD and DLB phenotypes are more aligned.

Relation to Yang et al 2019 and Frisoni et al 2004:
- All three of these studies investigated relationship between brain shape and Alzheimer's disease but with different morphometrics. Yang, Frisoni, and Burton assessed cortical thickness, gray matter density, and gray matter volume, respectively. Burton highlighted the temporal lobe (specifically hippocampal/amygdalar complex) as most affected by Alzheimer's, which both Yang's and Frisoni's findings largely supported.

Relation to Zarei et al 2012 and Blair et al 2019:
- All three of these studies investigated the relationship between brain shape and Parkinson's disease but with different morphometrics. Similar to the previous comparison above, Zarei, Blair, and Burton used cortical thickness, gray matter density, and gray matter volume, respectively.

## Conclusion
Different brain shape features can be used to distinguish neurodegenerative disorders such as Alheimer's and Parkisonson's disease. Some features are more effective at predicting disease in general, while some diseases are best predicted using certain features. In these papers, clear reductions in cortical thickness, gray matter density, and gray matter volume were observed in AD and PD compared to healthy controls as well as among disease stages (advanced AD and PD exhibited much more atrophy than earlier stages). Other features were not as effective such as cortical surface area.

The introduction and discussion sections of these papers are filled with contrasting claims on the relationships between brain shape and disease state. For example, many discuss how their study supports or refutes parts of Braak's hypothesis. Though MRI image processing and variable quantification is standardized, I do think these measurements are dependent upon environmental exposures the patients have had throughout their lifetimes and the number of samples in the datasets. Some of these papers list out very specific subregions that exhibit atrophy and others provide enumerable subregions. AD and PD showed obvious differences in regions with altered cortical thickness and gray matter density, but more comparisons among the different neurodegenerative disorders is required to know which features are the best features for particular disorders. Biology is not an exact science and thus variability and error will always exist. Though this may be unfamiliarity with neuroscience, I am a bit wary of the confidence in using some of these specific subregions in brain MRI images as biomarkers of AD or PD.

Gray matter: 
- Composed of neural cell bodies, axon terminals, and dendrites, as well as all nerve synapses
- Implicated in neurodegernative diseases such as Alzheimer's

White matter:
- Located beneath the gray matter cortex, composed of axon bundles (nerve fibers) connecting neurons in different brain regions into functional circuits
- Implicated in diseases related to nerve signal transit such as Multiple Sclerosis
